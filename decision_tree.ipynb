```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df  = pd.read_csv('Social_Network_Ads.csv')
```

```python
df
```

```python
x = df[['Age' , 'EstimatedSalary']]
```

```python
y= df['Purchased']
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(x,y)
```

```python
dt.classes_
```

```python
dt.decision_path
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(24,24))
plot_tree(dt , fontsize=8, feature_names=['age','sal'], filled=True, rounded=True, precision=2);
```

```python
new = pd.DataFrame(data=[[23,67000], [54,89090], [47,123000]], columns=x.columns)
```

```python
dt.predict(new)
```
